## What are Tensors?

A tensor is a specialized multi-dimensional array designed for mathematical and computational efficiency.

---

## Real-World Examples of Tensors

### 1. Scalars (0-Dimensional Tensors)

A scalar represents a single value.

#### Example
- Loss value after a forward pass in a neural network.
- Temperature, probability, or accuracy score.

#### Examples of Scalars

```python
5.0
-3.14
```

---

### 2. Vectors (1-Dimensional Tensors)

A vector represents a sequence or collection of values.

#### Example
- Word embeddings in Natural Language Processing (NLP).

#### Example Vector

```python
[0.12, -0.84, 0.33]
```

This can represent a word embedding generated using models like Word2Vec or GloVe.

---

### 3. Matrices (2-Dimensional Tensors)

A matrix represents tabular or grid-like data.

#### Example
- Grayscale images

Each value represents pixel intensity.

#### Example Matrix

```python
[
 [0, 255, 128],
 [34, 90, 180]
]
```

---

### 4. 3D Tensors (Colored Images)

A 3D tensor adds another dimension, commonly used for image channels.

#### Example
- RGB Images

#### Shape Format

```python
[width, height, channels]
```

#### Example

```python
[256, 256, 3]
```

Where:
- `256 × 256` → Image size
- `3` → RGB channels

---

### 5. 4D Tensors (Batches of RGB Images)

A 4D tensor adds the batch dimension.

#### Example
- Batch of colored images

#### Shape Format

```python
[batch_size, width, height, channels]
```

#### Example

```python
[32, 128, 128, 3]
```

Where:
- `32` → Number of images
- `128 × 128` → Image size
- `3` → RGB channels

---

### 6. 5D Tensors (Video Data)

A 5D tensor adds a time/frame dimension.

#### Example
- Video clips

#### Shape Format

```python
[batch_size, frames, width, height, channels]
```

#### Example

```python
[10, 16, 64, 64, 3]
```

Where:
- `10` → Number of video clips
- `16` → Frames per video
- `64 × 64` → Frame size
- `3` → RGB channels

---

## Why Are Tensors Useful?

### 1. Mathematical Operations

Tensors support efficient mathematical computations such as:
- Addition
- Multiplication
- Dot Product
- Matrix Multiplication
- Broadcasting

These operations are essential for neural networks.

---

### 2. Representation of Real-World Data

Different types of data can be represented using tensors.

#### Examples
- Images → 3D tensors
- Text → 2D or 3D tensors
- Audio → 1D or 2D tensors
- Videos → 5D tensors

#### Text Representation

```python
[sequence_length, embedding_size]
```

---

### 3. Efficient Computations

Tensors are optimized for hardware acceleration using:
- GPUs
- TPUs

This enables faster training and inference in deep learning models.

---

## Where Are Tensors Used in Deep Learning?

### 1. Data Storage

Training data such as:
- Images
- Text
- Audio
- Videos

are stored as tensors.

---

### 2. Weights and Biases

The learnable parameters of neural networks are stored as tensors.

#### Examples
- Weight matrices
- Bias vectors

---

### 3. Matrix Operations

Deep learning relies heavily on tensor operations such as:
- Matrix multiplication
- Dot products
- Broadcasting

---

### 4. Training Process

#### Forward Pass
Input tensors flow through the neural network layers.

#### Backward Pass
Gradients are computed as tensors during backpropagation.

These gradients are used to update model parameters.

# Comparison GPU vs CPU

````python
import torch
import time

# Define the size of the matrices
size = 10000  # Large size for performance comparison

# Create random matrices on CPU
matrix_cpu1 = torch.randn(size, size)
matrix_cpu2 = torch.randn(size, size)

# Measure time on CPU
start_time = time.time()

result_cpu = torch.matmul(matrix_cpu1, matrix_cpu2)  # Matrix multiplication on CPU

cpu_time = time.time() - start_time

print(f"Time on CPU: {cpu_time:.4f} seconds")

# Move matrices to GPU
matrix_gpu1 = matrix_cpu1.to('cuda')
matrix_gpu2 = matrix_cpu2.to('cuda')

# Measure time on GPU
start_time = time.time()

result_gpu = torch.matmul(matrix_gpu1, matrix_gpu2)  # Matrix multiplication on GPU

torch.cuda.synchronize()  # Ensure all GPU operations are complete

gpu_time = time.time() - start_time

print(f"Time on GPU: {gpu_time:.4f} seconds")
````

# Reshaping Tensors

````python
# Create a tensor
a = torch.ones(4, 4)

print(a)

# Reshape tensor
reshaped_tensor = a.reshape(2, 2, 2, 2)

print("\nReshaped Tensor:")
print(reshaped_tensor)

# Flatten tensor
flattened_tensor = a.flatten() # convert in one dimension

print("\nFlattened Tensor:")
print(flattened_tensor)
````

````python
# Create a random tensor with shape (2, 3, 4)
b = torch.rand(2, 3, 4)

print("Original Tensor:")
print(b)

# permute changes the order of dimensions
permuted_tensor = b.permute(2, 1, 0)

print("\nPermuted Tensor Shape:")
print(permuted_tensor.shape)

# unsqueeze adds a new dimension at the specified position
c = torch.rand(226, 226, 3)

unsqueezed_tensor = c.unsqueeze(0)

print("\nUnsqueezed Tensor Shape:")
print(unsqueezed_tensor.shape)

# squeeze removes dimensions with size 1
d = torch.rand(1, 20)

squeezed_tensor = d.squeeze()

print("\nSqueezed Tensor Shape:")
print(squeezed_tensor.shape)
````

# NumPy and PyTorch

````python
import torch
import numpy as np

# Create a PyTorch tensor
a = torch.tensor([1, 2, 3])

print("PyTorch Tensor:")
print(a)

# Convert PyTorch tensor to NumPy array
# numpy() converts tensor into NumPy array
b = a.numpy()

print("\nConverted to NumPy Array:")
print(b)

# Create a NumPy array
c = np.array([4, 5, 6])

print("\nNumPy Array:")
print(c)

# Convert NumPy array to PyTorch tensor
# from_numpy() converts NumPy array into PyTorch tensor
d = torch.from_numpy(c)

print("\nConverted to PyTorch Tensor:")
print(d)
````
